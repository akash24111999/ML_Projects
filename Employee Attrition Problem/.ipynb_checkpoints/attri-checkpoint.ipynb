```python
x_scaled  = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
rfe = RFE(model)
```

```python
fit = rfe.fit(x,y)
```

```python
print(fit.support_)
print(fit.ranking_)
```

```python
def feature_ranks(x,rank,support):
    feature_rank = pd.DataFrame()
    for i in range(x.shape[1]):
        new = pd.DataFrame({'Features':x.columns[i],'Rank':rank[i],'Selected':support[i]},index=[i])
        feature_rank = pd.concat([feature_rank,new])
    return feature_rank
```

```python
selected_features = feature_ranks(x,fit.ranking_,fit.support_)
selected_features
```

```python
rfe_select = selected_features[selected_features['Selected']==True]
```

```python
rfe_select
```

```python
x_scaled = x_scaled[rfe_select['Features'].values]
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3)
```

```python
lr = LinearRegression()
model = lr.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(pred,y_test)
```

```python
np.sqrt(mean_squared_error(y_test,pred))
```
